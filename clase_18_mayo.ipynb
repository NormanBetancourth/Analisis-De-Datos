```python
#Lab guiado

import folium

from folium.plugins import minimap
EscuelaInfo = folium.Map(location=(9.970709, -84.129109))
EscuelaInfo
```

```python
from folium.plugins import MiniMap
from branca.element import Figure
popuptext = "<br>Escuela de Informatica</br>"

my_map = folium.Map(location=[9.970709, -84.129109], zoom_start=16, tiles="stamenterrain")

folium.Marker(location=[9.970709, -84.129109], 
              popup=popuptext, icon=folium.Icon(color="orange", icon="ok-sign")).add_to(my_map)

folium.Circle(location=[9.970709, -84.129109], color="purple", weight=6, fill_color="red", fill_opacity=0.5, tooltip="Escuela de Informatica").add_to(my_map)

minimap = MiniMap(tile_layer="stamenterrain", position="bottomright")

my_map.add_child(minimap)

my_map.save("C:folium.html")

my_map

fig = Figure(width=1000, height=500)
fig.add_child(my_map)
fig
```
